# Correlation Map Function Definition

```python
# this defines the histogram for correlation map
def plot_histograms( df , variables , n_rows , n_cols ):
    fig = plt.figure( figsize = ( 16 , 12 ) )
    for i, var_name in enumerate( variables ):
        ax=fig.add_subplot( n_rows , n_cols , i+1 )
        df[ var_name ].hist( bins=10 , ax=ax )
        ax.set_title( 'Skew: ' + str( round( float( df[ var_name ].skew() ) , ) ) ) # + ' ' + var_name ) #var_name+" Distribution")
        ax.set_xticklabels( [] , visible=False )
        ax.set_yticklabels( [] , visible=False )
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()

def plot_distribution( df , var , target , **kwargs ):
    row = kwargs.get( 'row' , None )
    col = kwargs.get( 'col' , None )
    facet = sns.FacetGrid( df , hue=target , aspect=4 , row = row , col = col )
    facet.map( sns.kdeplot , var , shade= True )
    facet.set( xlim=( 0 , df[ var ].max() ) )
    facet.add_legend()

def plot_categories( df , cat , target , **kwargs ):
    row = kwargs.get( 'row' , None )
    col = kwargs.get( 'col' , None )
    facet = sns.FacetGrid( df , row = row , col = col )
    facet.map( sns.barplot , cat , target )
    facet.add_legend()

def plot_correlation_map( df ):
    corr = df.corr()
    _ , ax = plt.subplots( figsize =( 5 ,  5) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )

def describe_more( df ):
    var = [] ; l = [] ; t = []
    for x in df:
        var.append( x )
        l.append( len( pd.value_counts( df[ x ] ) ) )
        t.append( df[ x ].dtypes )
    levels = pd.DataFrame( { 'Variable' : var , 'Levels' : l , 'Datatype' : t } )
    levels.sort_values( by = 'Levels' , inplace = True )
    return levels

def plot_variable_importance( X , y ):
    tree = DecisionTreeClassifier( random_state = 99 )
    tree.fit( X , y )
    plot_model_var_imp( tree , X , y )
    
def plot_model_var_imp( model , X , y ):
    imp = pd.DataFrame( 
        model.feature_importances_  , 
        columns = [ 'Importance' ] , 
        index = X.columns 
    )
    imp = imp.sort_values( [ 'Importance' ] , ascending = True )
    imp[ : 10 ].plot( kind = 'barh' )
    print (model.score( X , y ))

    
# Correlation Map
plot_correlation_map( df.drop('species', axis=1) )
```

# Metrics 
***
 - **Accuracy**
 
        1. Predicted correctly/ total number of examples ((tp+tn)/total)
        2. Overall, how often is the classifier correct? 
        3. Sum of diagonals on the total



 - **Precision**
         1. When it predicts yes, how often is it correct? ((tp + fp) / predicted yes)
         2. Individual diagonal cell devided by the sum of the columns

 - **Recall (aka True Positive Rate or Sensitivity)**
        1. Predicted surviving that really survived / Total number that really survived
        2. When it's actually yes, how often does it predict yes? (tp/actual yes) 
        3. Individual diagonal cell devided by the sum of its row

### Confusion Matrix Plots for all Models

```python
normalize=False
title=None
cmap=plt.cm.Blues
# Compute confusion matrix

            
num_each = int(np.ceil(len(models)/2.0))


fig, ax = plt.subplots(nrows=3, ncols=3,figsize=(20, 20))
ii = 0
for row in ax:
    for my_ax in row:
        
        model = fit_model_list[ii][2][0]
        model_name = fit_model_list[ii][0]
        # fit
        y_pred = model.fit(X_train,y_train).predict(X_val)

        y_pred_new = np.array(list(map(lambda b: 2 if b=='virginica' else (1 if b =='versicolor' else 0),y_pred )))
        y_val_new = np.array(list(map(lambda b: 2 if b=='virginica' else (1 if b =='versicolor' else 0),y_val )))


        cm = confusion_matrix(y_val_new, y_pred_new)
        # Only use the labels that appear in the data

  
        classes = class_names[unique_labels(y_val_new, y_pred_new)]
        if normalize:
            cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

 
        
        im = my_ax.imshow(cm, interpolation='nearest', cmap=cmap)
        my_ax.figure.colorbar(im, ax=my_ax)
        # We want to show all ticks...
        my_ax.set(xticks=np.arange(cm.shape[1]),
            yticks=np.arange(cm.shape[0]),
            # ... and label them with the respective list entries
            xticklabels=classes, yticklabels=classes,
            title=model_name,
            # title=title,
            ylabel='True label',
            xlabel='Predicted label')



        # Rotate the tick labels and set their alignment.
        plt.setp(my_ax.get_xticklabels(), rotation=45, ha="right",
                rotation_mode="anchor")

        # Loop over data dimensions and create text annotations.
        fmt = '.2f' if normalize else 'd'
        thresh = cm.max() / 2.
        for i in range(cm.shape[0]):
            for j in range(cm.shape[1]):
                my_ax.text(j, i, format(cm[i, j], fmt),
                        ha="center", va="center",
                        color="white" if cm[i, j] > thresh else "black")
        ii+=1
fig.tight_layout()
plt.show()
```
